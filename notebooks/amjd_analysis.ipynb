```python
file = r"C:\Users\DSC-GDP075\Desktop\capstone\GDP6_CAPSTONE\data\bodymassindex_2024.xlsx"
df = pd.read_excel(file, header=1)
df.columns = ["Category", "Saudi_Male", "Saudi_Female", "Saudi_Total",
              "NonSaudi_Male", "NonSaudi_Female", "NonSaudi_Total"]
df.iloc[0, 0] = "Underweight"

df.head()

```

```python
df2 = pd.read_excel(file, header=1, sheet_name="Sheet2")  
df2.columns = ["Category", "Saudi_Male", "Saudi_Female", "Saudi_Total",
              "NonSaudi_Male", "NonSaudi_Female", "NonSaudi_Total"]
df2.iloc[0, 0] = "Underweight"
df2.head()
```

```python
plot_df = df.copy()
plot_df.iloc[:, 1:] = plot_df.iloc[:, 1:] * 100 

ax = plot_df.plot(
    x=plot_df.columns[0], 
    y="Saudi_Total",  
    kind="bar", 
    legend=False
)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')

plt.title("Body Mass Index (BMI) among children (2 - 14 years) by sex and nationality (Saudi)")
plt.ylabel("Percentage (%)")
plt.xlabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


plot_df = df.copy()
plot_df.iloc[:, 1:] = plot_df.iloc[:, 1:] * 100  

ax = plot_df.plot(
    x=plot_df.columns[0], 
    y="NonSaudi_Total", 
    kind="bar", 
    legend=False
)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom')

plt.title("Body Mass Index (BMI) among children (2 - 14 years) by sex and nationality (Non-Saudi)")
plt.ylabel("Percentage (%)")
plt.xlabel("Category")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
file = r"C:\Users\DSC-GDP075\Desktop\capstone\GDP6_CAPSTONE\data\SA - AU Country Table Export.xlsx"
df6 = pd.read_excel(file, header=None, sheet_name="Count")
df6.columns = ["Country", "2000", "2011","2024","2050"] 
df6 = df6.iloc[1:]


df6.head()
```

```python
df_melted = df6.melt(id_vars="Country", var_name="Year", value_name="Value")

for country in df6["Country"].unique():
    data = df_melted[df_melted["Country"] == country]
    plt.plot(data["Year"], data["Value"], marker='o')

plt.xlabel("Year")
plt.ylabel("GDP / Value")
plt.title("People with diabetes, in 1,000s")
plt.legend(df6["Country"].unique())
plt.grid(True)
plt.show()
```

```python
file = r"C:\Users\DSC-GDP075\Desktop\capstone\GDP6_CAPSTONE\data\diabetes_diabetes.xlsx"
df3 = pd.read_excel(file, header=None, sheet_name="Sheet1")
df3.columns = ["Year", "Males", "Females"] 
df3 = df3.iloc[1:]


df3.head()
```

```python
df3["Year"] = pd.to_timedelta(df3["Year"])
df3["Males"] = pd.to_numeric(df3["Males"])
df3["Females"] = pd.to_numeric(df3["Females"])

plt.figure()
plt.plot(df3["Year"], df3["Males"])
plt.plot(df3["Year"], df3["Females"])

plt.xlabel("Year")
plt.ylabel("Incidence")
plt.legend(["Males", "Females"])
plt.title("Diabetes Trends Over Years (Australia)")
plt.tight_layout()
plt.show()
```
